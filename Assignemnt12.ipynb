```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
```

```python
df = pd.read_csv('Salary.csv')
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['YearsExperience'], df['Salary'], color='blue', label='Actual Data')
plt.title('Years of Experience vs Salary')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
X = df[['YearsExperience']].values
y = df['Salary'].values
```

```python
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)
```

```python
model = LinearRegression()
model.fit(X_poly, y)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
```

```python
y_pred = model.predict(X_poly)
```

```python
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)
```

```python
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"R² Score: {r2:.4f}")
```

```python
X_grid = np.linspace(min(X)[0], max(X)[0], 100).reshape(-1, 1)
X_grid_poly = poly.transform(X_grid)
y_grid_pred = model.predict(X_grid_poly)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X_grid, y_grid_pred, color='red', label='Polynomial Regression')
plt.title('Polynomial Regression Curve')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.grid(True)
plt.show()
```
